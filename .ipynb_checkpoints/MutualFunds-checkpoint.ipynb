## Utilities

```python
def sample(a_dictionary, a_number): 
    _ = {}
    for k1 in random.sample(list(a_dictionary.keys()) , min(len(list(a_dictionary.keys())), a_number)):
        _[k1] = a_dictionary[k1]
    return _
```

## Scheme Details

### Fetching

```python
# Opening JSON file
f = open('codes.json')
 
# returns JSON object as 
# a dictionary
scheme_codes_all = json.load(f)
```

```python
## Limit values
## Remove this cell to run for all

sample_size = None #100

if sample_size is not None:
    scheme_codes = random.sample(scheme_codes_all, sample_size)
else:
    scheme_codes = scheme_codes_all
```

```python
scheme_df = None

class Schemes:

    def __init__(self, df):
        self.df = df

    def _get_scheme_code(self, scheme_code_row):
        return list(scheme_code_row)[0]

    def _get_scheme_details(self, scheme_code):
        try:
            scheme_details = mf.get_scheme_info(scheme_code, as_json=False)
        except Exception as e:
            print('Exception while fetching Scheme Code {:s} - {:s}'.format(scheme_code, repr(e)))
            return {'scheme_code': scheme_code}
            
        scheme_details['schemeCode'] = scheme_code
        return scheme_details

    def _get_scheme_df(self, scheme_details):
        return pd.DataFrame(scheme_details, index=[scheme_details['scheme_code']])

    def get_all_schemes(self, scheme_codes):
        for scheme_code_row in scheme_codes:
            scheme_code = self._get_scheme_code(scheme_code_row)
            scheme_details = self._get_scheme_details(scheme_code)

            if self.df is None:
                self.df = self._get_scheme_df(scheme_details)
            else:
                df = self._get_scheme_df(scheme_details)
                self.df = pd.concat([self.df, df])

            print("Fetched Scheme Code: " + scheme_code)
        return self.df

Schemes = Schemes(scheme_df)
scheme_df = Schemes.get_all_schemes(scheme_codes)
```
